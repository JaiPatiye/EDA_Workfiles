## KDE Plots

```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing data
data = pd.read_csv('churn_prediction.csv')
```

```python
data.dtypes, data.shape
```

```python
# extracting age of people of who churned
churn_age = data['age'][data['churn']==1]

#checking
churn_age
```

```python
# plotting histogram
plt.figure(dpi = 140)
plt.hist(churn_age)
plt.xlabel("Age")
plt.ylabel('Frequency')
plt.title("histogram : Age of people who churned")
plt.show()
```

```python
# plotting histogram with 5 bins
plt.figure(dpi = 140)
plt.hist(churn_age, bins=5)
plt.xlabel("Age")
plt.ylabel('Frequency')
plt.title("histogram : Age of people who churned")
plt.show()
```

```python
# plotting histogram with 25 bins
plt.figure(dpi = 140)
plt.hist(churn_age, bins = 25)
plt.xlabel("Age")
plt.ylabel('Frequency')
plt.title("histogram : Age of people who churned")
plt.show()
```

```python
# plotting KdE plot using Seaborn
plt.figure(dpi = 140)
sns.kdeplot(churn_age, shade = True)
plt.xlabel("Age")
plt.ylabel('Density')
plt.title("KDE : Age of people who churned")
plt.show()
```

```python
# calculating points of mean, median and standard deviation
mean = churn_age.mean()
median = churn_age.median()
stdev = churn_age.mean()- churn_age.std(), churn_age.mean()+churn_age.std() 

# checking
mean, median, stdev
```

```python
# plotting KDE plot with descriptives
plt.figure(dpi = 140)
sns.kdeplot(churn_age, shade = True)
sns.scatterplot( [mean], [0], color='red', label='mean')
sns.scatterplot( [median], [0], color='green', label='median')
sns.lineplot( stdev, [0,0], color='black', label='standard deviation')
plt.xlabel("Age")
plt.ylabel('Density')
plt.title("KDE Plot with descriptives : Age of people who churned")
plt.show()
```

## Prepare a kde plot for continuous variables like current balance, average balance in the present quarter and current month debit/credit

```python
# extracting age of people of who churned
churn_current_balance = data['previous_month_debit'][data['churn']==1]

#checking
churn_current_balance
```

```python
# calculating points of mean, median and standard deviation
mean = churn_current_balance.mean()
median = churn_current_balance.median()
stdev = churn_current_balance.mean()- churn_current_balance.std(), churn_current_balance.mean()+churn_current_balance.std() 
max_balance = churn_current_balance.max()
# checking
mean, median, stdev, max_balance

```

```python
# plotting KDE plot with descriptives
plt.figure(dpi = 140)
sns.kdeplot(churn_current_balance, shade = True)
sns.scatterplot( [mean], [0], color='red', label='mean')
sns.scatterplot( [median], [0], color='green', label='median')
sns.lineplot( stdev, [0,0], color='black', label='standard deviation')
plt.xlabel("Current Balance")
plt.ylabel('Density')
plt.title("KDE Plot with descriptives : Current Balance of people who churned")
plt.show()
```
