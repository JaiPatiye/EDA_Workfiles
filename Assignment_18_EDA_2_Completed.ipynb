### 2. What kind of distribution is followed by Applicant Income ?

```python
plt.figure(dpi = 140)
plt.hist(data['ApplicantIncome'])
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.title("Distribution for Applicant Income")
plt.show()
print("Right skewed Distribution is followed by applicant income")
```

### 3. What kind of distribution is followed by CoApplicant Income ?

```python
plt.figure(dpi = 140)
plt.hist(data["CoapplicantIncome"])
plt.xlabel("Coapplicant Income")
plt.ylabel("Frequency")
plt.title("Distribution of Coapplicant's Income")
plt.show()
print("Right skewed ditribution")
```

### 4. What kind of distribution is followed by LoanAmount ?

```python
plt.figure(dpi = 140)
plt.hist(data['LoanAmount'])
plt.xlabel('Loan Amount')
plt.ylabel('Frequency')
plt.title('Distribution of loan amount')
plt.show()
print("Right skewed distribution")
```

### 5. Plot KDE plot with descriptive statistics for Loan Amount ?

```python
loan_amount = data.LoanAmount
loan_amount_mean = loan_amount.mean()
loan_amount_median = loan_amount.median()
loan_amount_stdev = loan_amount.std()



plt.figure(dpi = 140)
sns.kdeplot(loan_amount , shade = True)
sns.scatterplot([loan_amount_mean] , [0] , color = "red" , label = "Mean" )
sns.scatterplot([loan_amount_median] , [0] , color = "green" , label = "Median")
sns.scatterplot([loan_amount_stdev] , [0] , color = "black" , label = "Standard Deviation")
plt.xlabel("Loan Amount")
plt.ylabel("Frequency")
plt.title("KDE Plot with descriptive statistics")
plt.show()
```

### 6. Observe QQ plot for Loan Amount, Applicant Income and CoApplicant Income, derive your insights from them ?

```python
applicant_income = data.ApplicantIncome
coapplicant_income = data.CoapplicantIncome
sm.qqplot(applicant_income , line = 's')
py.show()

sm.qqplot(coapplicant_income , line = 's')
py.show()
```
